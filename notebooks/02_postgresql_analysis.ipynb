```python
import psycopg2
import pandas as pd
import sqlalchemy
```

```python
DB_USER = "postgres"
DB_PASSWORD = "pricass00"
DB_HOST = "localhost"
DB_PORT = "5432"
DB_NAME = "ecommerce_ai_analysis"
```

```python
from sqlalchemy import create_engine

engine = create_engine(
    f"postgresql+psycopg2://{DB_USER}:{DB_PASSWORD}@{DB_HOST}:{DB_PORT}/{DB_NAME}"
)

print("PostgreSQL connected successfully")
```

```python
import pandas as pd

df = pd.read_csv("cleaned_data.csv")

df.to_sql(
    name="ecommerce_sessions",
    con=engine,
    if_exists="replace",   
    index=False
)

print("Data successfully loaded into PostgreSQL")
```

```python
query = "SELECT COUNT(*) FROM ecommerce_sessions;"
result = pd.read_sql(query, engine)
result
```
